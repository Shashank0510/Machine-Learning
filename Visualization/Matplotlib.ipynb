## Introduction

### What is Matplotlib ?

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

### Importing Matplotlib 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
```

## Plots

### Scatter Plot

```python
plt.scatter(x,y,c='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot Graph in 2D")
plt.savefig("Scatterplot.png")
```

### Plot

```python
y=x*x
plt.plot(x,y,'r--*',markersize=12)
plt.xlabel("X")
plt.ylabel("X * X")
plt.title("Graph for X^2 function")
```

### Sub Plots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.ylabel("y")
plt.xlabel("x")
plt.subplot(2,2,2)
plt.plot(x,y,'b*')
plt.subplot(2,2,3)
plt.plot(x,y,c='g')
plt.subplot(2,2,4)
plt.plot(x,y,c='y')
```

### Sin Graph

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

### Cos Graph

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
```

### Sub Plot of Sin and Cos Graph

```python
plt.subplot(1,2,1)
plt.plot(x,np.sin(x),'go')
plt.subplot(1,2,2)
plt.plot(x,np.cos(x),'b-')
```

### Bar Graph

```python
x=[2,8,10]
y=[11,16,19]
x2=[3,9,11]
y2=[6,11,15]
plt.bar(x,y,color=['green'])
plt.bar(x2,y2)
plt.title("Bar graph")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

### Histogram

```python
x=np.array([12,34,56,76,98,13,14,15,18,32,53,73,25])
plt.hist(x,bins=10)
plt.title("Histogram")
plt.show()
```

### Box Plot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

### Pie Chart

```python
labels='Python','Java','C++','Ruby'
sizes=[215,130,245,210]
colors=['red','blue','green','gold']
explode=(0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True,autopct='%1.1f%%')
plt.show()
```
