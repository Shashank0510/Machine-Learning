## What is Seaborn ?

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

### Importing Seaborn 

```python
import seaborn as sns
```

### Loading data Set
Note:Seaborn has built in dataset named -**Tips**

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

### Numerical Plots

#### Correlation

Correlation is a statistic that measures the degree to which two variables move in relation to each other.

```python
df.corr()
```

#####  Heatmap

```python
sns.heatmap(df.corr())
```

##### Joint Plot

**Seabornâ€™s jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.**

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

##### Pairplot

**To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="smoker")
```

```python
df['smoker'].value_counts()
```

##### Distplot

**It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.**

```python
sns.distplot(df['tip'], kde=True, bins=10)
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

### Categorical Plots

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
sns.barplot(x="smoker",y="total_bill",data=df)
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
```

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot('total_bill','day',data=df,hue="smoker")
```

```python
sns.violinplot('total_bill','day',data=df,hue="smoker")
```
