### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
sales2=sales2.rename(columns={'Title':'Book title','Units sold':'Number sold','List price':'Sales price','Royalty':'Royalty paid'})
```

```python
s = sales2.copy()
t = sales2['Book title']
t = t.where(t.str.endswith(')')).str.split().str[-1].str.strip('()')
s['Currency'] = t.fillna(method='bfill')
s = s[s['Sales price'].notnull()]
s
```

```python
df=pd.concat([s,sales1],axis=0)
df
```

```python
df.groupby('Currency').size()
```

```python
t = df
t = t.assign(Total_Paid=t['Sales price'] * t['Number sold'])
t.groupby(['Book title', 'Currency'])[['Total_Paid']].sum()
```
